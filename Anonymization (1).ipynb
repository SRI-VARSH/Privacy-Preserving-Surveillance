```python
from insightface.app import FaceAnalysis

ctx_id = 0 if torch.cuda.is_available() else -1

app = FaceAnalysis(name='buffalo_l')

# 🔥 IMPORTANT FIX
app.prepare(ctx_id=ctx_id, det_size=(640, 640))

print("✅ InsightFace initialized")
```

```python
DONOR_PATH = "/content/donor1.jpg"

donor_img = cv2.imread(DONOR_PATH)
assert donor_img is not None, f"❌ Cannot load donor image: {DONOR_PATH}"

donor_img = cv2.resize(donor_img, (512, 512))

donor_faces = app.get(donor_img)
assert len(donor_faces) > 0, "❌ No face detected in donor image"

donor_face = donor_faces[0]

print("✅ Donor face loaded once and ready")
```

**SWAP FUNCTION**

```python
def is_same_face(b1, b2, threshold=60):
    x1, y1, x2, y2 = map(int, b1)
    x1_, y1_, x2_, y2_ = map(int, b2)

    cx1, cy1 = (x1 + x2) // 2, (y1 + y2) // 2
    cx2, cy2 = (x1_ + x2_) // 2, (y1_ + y2_) // 2

    return abs(cx1 - cx2) < threshold and abs(cy1 - cy2) < threshold


def swap_face(frame, bbox, donor_face):
    # 🔍 Detect all faces in full frame
    faces = app.get(frame)

    if len(faces) == 0:
        print("⚠️ No faces detected in full frame")
        return frame

    matched = False

    for face in faces:
        fx1, fy1, fx2, fy2 = face.bbox.astype(int)

        # 🔗 Match tracker bbox with detected face
        if is_same_face((fx1, fy1, fx2, fy2), bbox):
            matched = True

            try:
                frame = swapper.get(
                    frame,
                    face,
                    donor_face,
                    paste_back=True
                )
                print("✅ Swap successful")

            except Exception as e:
                print("❌ Swap failed:", type(e).__name__, e)

    if not matched:
        print("⚠️ No matching face found for bbox:", bbox)

    return frame
```

**MAIN FUNCTION**

```python
def run_anonymization_pipeline():
    print("🚀 Running embedding pipeline...")
    embedding_results = run_embedding_pipeline(DATASET_PATH)

    frame_files = sorted(os.listdir(DATASET_PATH))
    MAX_FRAMES = 300

    for frame_data in embedding_results[:MAX_FRAMES]:
        frame_id = frame_data["frame_id"]

        if frame_id >= len(frame_files):
            continue

        frame_path = os.path.join(DATASET_PATH, frame_files[frame_id])
        frame = cv2.imread(frame_path)

        if frame is None:
            print(f"❌ Failed to read frame {frame_id}")
            continue

        print(f"\n🧠 Processing frame {frame_id}")

        for track in frame_data["tracks"]:
            bbox = track["bbox"]

            # 🔍 DEBUG bbox format
            if len(bbox) == 4:
                x1, y1, x2, y2 = bbox

                # ⚠️ If it's (x, y, w, h), convert it
                if x2 < x1 or y2 < y1:
                    print("⚠️ Detected (x,y,w,h) format — converting")
                    x, y, w, h = bbox
                    bbox = (x, y, x + w, y + h)

            print("➡️ Processing bbox:", bbox)

            frame = swap_face(frame, bbox, donor_face)

        out_path = os.path.join(OUTPUT_PATH, f"{frame_id:05d}.jpg")
        cv2.imwrite(out_path, frame)

    print("✅ Anonymization complete")
```

```python
run_anonymization_pipeline()
```
